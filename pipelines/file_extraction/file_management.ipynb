# Unzipping files

```python
import py7zr
import os
import numpy as np
import pandas as pd
```

```python
def extract_7z(file_path, output_path):
    with py7zr.SevenZipFile(file_path, mode='r') as z:
        z.extractall(path=output_path)
```

```python
raw_data_root = "/home/mp422/data/raw"
files_to_extact = os.listdir(raw_data_root)
paths_to_extact = ["/".join([raw_data_root,f]) for f in files_to_extact]
```

```python
output_root = "/home/mp422/data/input"
for input_path in paths_to_extact:
    print(f"Processing {input_path}")
    if not os.path.exists(output_root):
        os.makedirs(output_root)
    extract_7z(input_path, output_root)
    print(f"Extracted {input_path} to {output_root}")
```

```python
extracted_files = os.listdir(output_root)
```

```python
extracted_files
```

```python
dates = []
file_types = []
for extracted_file in extracted_files:
    dates.append(extracted_file.split("_")[1])
    file_types.append(extracted_file.split("_")[4])
```

```python
df = pd.DataFrame({"date":dates, "file_type":file_types})
```

```python
df.groupby(["date","file_type"]).agg({"file_type":"count"})
```
